```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tk = Tokenizer()
```

```python
tk.fit_on_texts([data])
```

```python
len(tk.word_index)
```

```python
squence=[]
for x in data.split('\n'):
    tokenised_sentence = tk.texts_to_sequences([x])[0]
    for i in range(1,len(tokenised_sentence)):
        squence.append(tokenised_sentence[:i+1])
```

```python
squence
```

```python
max_len = max([len(x) for x in squence])
max_len
```

```python
from tensorflow.keras.utils import pad_sequences
input_squence = pad_sequences(squence,maxlen = max_len, padding ='pre')
```

```python
X = input_squence[:,:-1]
X.shape
```

```python
tk.word_index
```

```python
y = input_squence[:,-1]
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 255)`
```

```python
y
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM, Bidirectional
```

```python
model = Sequential()
model.add(Embedding(255, 100, input_length=23))
model.add(Bidirectional(LSTM(150)))
model.add(Dense(255,activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=50)
```
